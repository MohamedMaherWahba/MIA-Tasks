```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('shrink_ray_dataset.csv')
x = data[["Power"]]
y = data[["Shrinkage"]]
```

```python
model = LinearRegression()
model.fit(x, y)
```

```python
y_pred = model.predict(x)
```

```python
plt.scatter(x, y, color = 'blue', label = "Collected")
plt.plot(x, y_pred, color = 'red', label = "Predicted")
plt.title('Power vs Shrinkage')
plt.xlabel('Power')
plt.ylabel('Shrinkage')
plt.legend()
plt.show()
```

```python
required_x =(85 - model.intercept_) / model.coef_
print(required_x)
```
